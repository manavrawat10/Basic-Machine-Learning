```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot([1,2,3], [5,6,4])
plt.show()
```

```python
x=[1,2,3]
y=[4,5,3]
plt.plot(x, y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("X vs Y Graph")
plt.show()
```

```python
x=[1,2,3]
y=[4,5,3]
x2=[1,2,3]
y2=[8,7,5]
plt.plot(x, y,label='First line')
plt.plot(x2,y2,label='Second Line')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("X vs Y Graph")
plt.legend()
plt.show()
```
